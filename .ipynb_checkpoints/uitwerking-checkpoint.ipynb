### Van alle data, naar de gemiddelde verkoopprijs per provincie
We beginnen met het minimaliseren en hernoemen van de columns.

```python
df_verkoopprijs_nl.drop(['ID'], axis=1, inplace=True)
df_verkoopprijs_nl.rename(columns={'RegioS':'Regio', 'Perioden':'Jaar', 'GemiddeldeVerkoopprijs_1':'Gemiddelde Verkoopprijs'}, inplace=True)
df_verkoopprijs_nl.head()
```

Ik weet nog niet zeker of dit nodig is, maar we willen straks de datasets combineren. Voor het gemak beginnen we onze data op het jaar 2008, aangezien we daarvoor in de welbekende bubble leefden.

```python
df_verkoopprijs_provincies['Jaar'] = df_verkoopprijs_provincies['Jaar'].astype(int) # Het datatype zetten we hier op int, zodat we in de volgende regel een integer kunnen gebruiken in de formule.
indexNames = df_verkoopprijs_provincies[ df_verkoopprijs_provincies['Jaar'] < 2008 ].index
df_verkoopprijs_provincies.drop(indexNames , inplace=True)

df_verkoopprijs_provincies.info()
```

## Gemiddelde prijsindex
De gemiddelde verkoopprijs kan een andere ontwikkeling laten zien dan de Prijsindex Bestaande Koopwoningen (PBK). De gemiddelde koopsom is geen indicator voor de prijsontwikkeling van koopwoningen. De gemiddelde koopsom weerspiegelt namelijk de gemiddelde prijs van woningen die verkocht zijn in een specifieke periode. Doordat elke periode andere woningen worden verkocht, wordt er geen rekening gehouden met eventuele verschillende kenmerken van de verkochte woningen. 

> Een voorbeeld om uit te leggen welke problemen dit met zich meebrengt: Stel in februari van een jaar zijn voornamelijk grote grachtenpanden met tuin verkocht, terwijl in maart veel kleine tussenwoningen van eigenaar wisselden. Hierdoor zal de gemiddelde koopsom in februari hoger liggen dan in maart maar dit wil niet zeggen dat de huizenprijzen zijn gestegen.

Meer achtergrondinformatie kan worden verkregen op: https://www.cbs.nl/nl-nl/achtergrond/2008/21/waarom-de-gemiddelde-koopsom-geen-huizenprijsindicator-is.

```python
dfs = [df_verkoopprijs_provincies, df_woz_provincies]

df_verkoop_indexed = df_verkoopprijs_provincies.set_index(['Regio', 'Jaar']) 
df_woz_indexed = df_woz_provincies.set_index(['Regio', 'Jaar'])
```

We kunnen met behulp van `apply()` en `pd.to_numeric` ervoor zorgen dat het datatype van kolom Jaar naar int veranderd. Dit kunnen we controleren met `.dtypes`.

```python
df_verkoopprijs_provincies[['Jaar']] = df_verkoopprijs_provincies[['Jaar']].apply(pd.to_numeric, errors='coerce') 
df_verkoopprijs_provincies.dtypes
```

Nu gaan we de verkoopprijzen en woz samenvoegen met behulp van `merge()`. `how='inner'` zorgt ervoor dat het vergelijkbaar is met een SQL inner join. Gebruik kruising van sleutels uit beide frames. Met `on` bepalen we de namen op kolom- of indexniveau om mee te doen. Deze moeten in beide DataFrames te vinden zijn.

```python
df_prijsindex = df_verkoop_indexed.merge(df_woz_indexed, how='inner', on=['Regio', 'Jaar'])
display(df_prijsindex)
```

Nu ze er beide instaan kunnen we de Prijsindex toevoegen met behulp van een simpele formule.

```python
df_prijsindex['Prijsindex'] = df_prijsindex['Gemiddelde Verkoopprijs'] / df_prijsindex['Gemiddelde WOZ']
df_prijsindex.head()
```

Wanneer de PBK gegevens van het CBS worden bekeken, is te zien dat onze gegevens hier iets van afwijken. Dit heeft de volgende redenen: 
- Vanaf 21 februari 2013 zijn de gewichten gebaseerd op de waarde van de woningvoorraad en niet meer op de som van de verkopenprijzen van de woningen die zijn verkocht in een recent jaar. 
- Bij de publicatie op 21 februari 2014 zijn de gewichten verder aangescherpt. De nieuwe gewichten zijn gebaseerd op de voorraad koopwoningen in plaats van op de totale woningvoorraad.

Onze berekening is daarom iets banaler en is enkel te gebruiken voor deze eindopdracht.

### Data Hypotheekrente ordenen
Wat betekend hypotheekrente eigenlijk? Een hypotheek is in feite een lening voor je woning. Over deze lening betaal je rente: de hypotheekrente. De hoogte van de rente bepaalt deels hoeveel jij per maand aan hypotheeklasten betaalt. Een lagere hypotheekrente leidt dus tot lagere maandlasten.

Het is voor ons een interessant gegeven, omdat de hoogte van de hypotheekrente dus ook weer bepaald hoe hoog, of laag, je vaste lasten zijn. Een lage hypotheekrente staat dus ongeveer gelijk aan een hogere mogelijke lening en dus duurdere woningen.

### Prijsindex
Nu gaan we de prijsindex plotten. Dit doen we op precies dezelfde manier, maar dan met een andere DataFrame.

```python
fig_dims = (16, 10)
fig, ax = plt.subplots(figsize=fig_dims)

sns.lineplot(data=df_prijsindex, x = 'Jaar', y = 'Prijsindex', hue='Regio', ax = ax, ci=0)
```

Ik denk dat we dit beter kunnen bekijken als we het gemiddelde van Nederland bekijken. Dit kunnen we door het gebruik van `mean()` op alle kolommen. Voordat we dit doen kunnen we met behulp van `transpose()` de kolom kantelen en gelijk een kolom Gemiddelde toevoegen.

```python
df_prijsindex_clean = df_prijsindex.drop(['Gemiddelde Verkoopprijs', 'Gemiddelde WOZ'], axis=1)

df_pi_stacked = df_prijsindex_clean.unstack()
df_pi_stacked = df_pi_stacked.transpose()

df_pi_stacked['Gemiddelde'] = df_pi_stacked.mean(numeric_only=True, axis=1)
df_pi_stacked.head()
```

## Deel IV: visualisaties van Nederland met Folium

Zoals gewoonlijk beginnen we met het importeren van de gebruikte package, Folium.

Folium is een Python bibliotheek die wordt gebruikt voor het visualiseren van geospatiale gegevens. Het is gemakkelijk te gebruiken en toch een krachtige bibliotheek.

Eens kijken waar we mee te maken hebben.

```python
df_verkoopprijs_provincies_pivot.head()
```

Wij willen de verkoopwaarde-stijging in het laatste jaar weergeven. Om dit makkelijk te doen kantelen we de tabel weer met `transpose()` en pakken we alleen 2020.

```python
df_prijs_lastyear = df_verkoopprijs_provincies_pivot[2020].transpose()

df_prijs_lastyear = df_prijs_lastyear.to_frame()

df_prijs_lastyear.reset_index(inplace=True)
df_prijs_lastyear.head(12)
```

Omdat we een `geojson` gebruiken, waarbij de namen overeen moeten komen met onze tabel gaan we de provincies strippen van de echte Friese taal en de "(PV)" tag. Goed opletten dat we ook de spatie meenemen vóór de tag.

```python
df_prijs_lastyear['Regio'] = df_prijs_lastyear['Regio'].map(lambda x: x.rstrip(' (PV)'))
df_prijs_lastyear['Regio'] = df_prijs_lastyear['Regio'].replace({'Fryslân': 'Friesland'})
df_prijs_lastyear.head(12)
```

Om een mooie render te maken gebruiken we Choropleth. Dit is tot nu toe een van mijn favoriete analyse-tools om kaarten en plattegronden mooi (en gemakkelijk) weer te geven.

De location kun je via google o.d. opzoeken, longi- en latitude, waarbij het zoomniveau bepaald van hoe ver je dit ziet. Het `geojson` bestand en je gewenste DataFrame zet je er ook in. Zorg dat de data klopt! Dan moeten we letten op `key_on`, daar voeren we de locatie in waar de plaatsnamen te vinden zijn in onze `geojson`. Dat kan wel eens wat puzzelen zijn. De rest van de instellingen zijn voornamelijk om de weergave van de plot te veranderen.

Dan nu, plotten maar!!

```python
provincies_geo = r'provinces.geojson'

nl_map = folium.Map(location=[52.12, 5.16], zoom_start=7)

folium.Choropleth(
    geo_data=provincies_geo,
    data=df_prijs_lastyear,
    columns=['Regio', 2020],
    key_on='feature.properties.name',
    fill_color='BuPu',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='Stijging verkoopprijzen 2020',
).add_to(nl_map)

nl_map
```

Dit zit er goed uit. Je kunt hier goed zien dat in 2020 de woningen in de goedkopere provincies in trek waren.

Ter vergelijking kun je [deze link](https://www.cbs.nl/nl-nl/nieuws/2021/03/prijzen-bestaande-koopwoningen-stijgen-in-2020-door-naar-recordniveau) volgen, hier heeft het CBS de prijsindex weergegeven van 2020. Dit is dus net iets anders van de verkoopprijs met redenen die we eerder al hebben besproken, maar de afbeelding komt goed overeen.

## Uitdaging voor mezelf
Een leuke uitdaging is om onze data te vergelijken met de data in het volgende artikel door de afbeelding Gemiddelde WOZ-waarde per gemeente na te bootsen.

https://www.gemeente.nu/bedrijfsvoering/belastingen/cbs-gemiddelde-woz-waarde-hoogste-ooit

Omdat we de gemeenten nodig hebben moeten we deze uit de originele df_woz_nl DataFrame halen, die we aan het begin van de eindopdracht hebben aangemaakt.

```python
df_woz_lastyear = df_woz_nl

df_woz_lastyear['Jaar'] = df_woz_lastyear['Jaar'].astype(int)
indexNames = df_woz_lastyear[ df_woz_lastyear['Jaar'] < 2020 ].index
df_woz_lastyear.drop(indexNames, inplace=True)

df_woz_lastyear = df_woz_lastyear.reset_index()
df_woz_lastyear.head()
```

Dan nu, plotten maar!!

```python
townships_geo = r'townships.geojson'

nl_map = folium.Map(location=[52.12, 5.16], zoom_start=7, tiles='OpenStreetMap')

folium.Choropleth(
    geo_data=townships_geo,
    data=df_woz_lastyear,
    columns=['Regio', 'Gemiddelde WOZ'],
    key_on='feature.properties.name',
    fill_color='Paired',
    fill_opacity=0.7,
    line_color='white',
    line_opacity=0.3,
    legend_name='Gemiddelde WOZ 2020',
    bins = 8 # Door het veranderen van deze waarde kun je hoeveelheid bins aanpassen. 
             # Waardoor de afbeelding meer of minder verschillende waarden weergeeft.
).add_to(nl_map)

nl_map
```

Los van de zwarte vlakken, waar een mismatch is tussen namen in de geojson en dataset, een succes!
